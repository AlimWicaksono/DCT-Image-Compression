```python
import numpy as np
import dct
import zigzag
import cv2
import matplotlib.pyplot as plt

from math import log10
from runlength import encode, decode
from jpeg_compression import *
from PIL import Image
%matplotlib inline
np.set_printoptions(formatter={'float': lambda x: "{0:8.3f}".format(x)})
```

```python
def histogram(img):
    color = ('b','g','r')
    for channel,col in enumerate(color):
        histr = cv2.calcHist([np.array(img)],[channel],None,[256],[0,256])
        plt.plot(histr,color = col)
        plt.xlim([0,256])
    plt.show()
    pass


def load_image(infilename):
    img = Image.open(infilename)
    img.load
    fig = plt.figure(figsize=(11, 5))
    fig.add_subplot(1, 2, 1)
    plt.imshow(np.asarray(img))
    fig.add_subplot(1, 2, 2)
    histogram(img)
    return np.array(img, dtype='float') - 127


def save_image(array, outfilename):
    img_arr = np.array(np.clip(np.round(array,0) + 127, 0, 255), dtype='uint8')
    img = Image.fromarray(img_arr, 'RGB')
    img.save(outfilename)
    fig = plt.figure(figsize=(11, 5))
    fig.add_subplot(1, 2, 1)
    plt.imshow(np.asarray(img))
    fig.add_subplot(1, 2, 2)
    histogram(img)
    return img_arr
```

```python
INPUT_FILE = './images/input/lena.tif'
COMPRESSED_FILE = './images/compressed/lena.npy'
OUTPUT_FILE = './images/output/lena rec.tif'
```

```python
img = load_image(INPUT_FILE)
[h, w, c] = img.shape
```

```python
yuv_img = rgb_to_yuv(img)
yuv_bands = split_channel(yuv_img)
print(img[0][0], yuv_img[0][0])
```

```python
yuv_blocks = [slice(band) for band in yuv_bands]
print(yuv_blocks[0][0][0])
```

```python
yuv_dct = [[[dct.transform(block) for block in row]
            for row in band] for band in yuv_blocks]  
print(yuv_dct[0][0][0])
```

```python
yuv_dct_q = np.array(
            [quantize(yuv_dct[0], channel='lum'),
             quantize(yuv_dct[1], channel='cr'),
             quantize(yuv_dct[2], channel='cr')])
print(yuv_dct_q[0][0][0])
```

```python
zigzag_ordered = [[[zigzag.scan(block) for block in row]
                   for row in band] for band in yuv_dct_q]  
print(zigzag_ordered[0][0][0])
```

```python
zigzag_ordered = np.array(zigzag_ordered, dtype='int8')
encoded = encode(zigzag_ordered, dtype='int8')
np.save(COMPRESSED_FILE, encoded)
c = h * w * c * 8.0
c1 = len(encoded) * 2 * 8.0
print(c, c1, 1 - c/c1)
```

```python
decoded = decode(np.load(COMPRESSED_FILE))
zigzag_arange = [[[zigzag.arange(block) for block in row]
                  for row in band] for band in decoded]
print(zigzag_arange[0][0][0])
```

```python
yuv_dct_deq = [dequantize(zigzag_arange[0], channel='lum'), 
               dequantize(zigzag_arange[1], channel='cr'),
               dequantize(zigzag_arange[2], channel='cr')]
print(yuv_dct_deq[0][0][0])
```

```python
yuv_idct = [[[dct.inverse(block) for block in row]
             for row in band] for band in yuv_dct_deq]
print(yuv_idct[0][0][0])
```

```python
yuv_rec = np.dstack([merge(band) for band in np.array(yuv_idct)])
rgb_rec = yuv_to_rgb(yuv_rec)
print(yuv_rec[0][0], rgb_rec[0][0])
```

```python
img_rec = save_image(rgb_rec, OUTPUT_FILE)
```

```python
mse = sum(sum(sum((img - img_rec) ** 2))) / (h * w * c)
psnr = 10 * log10((255 ** 2) / mse)
print(mse, psnr)
```
